## 1. Introducción

La interacción gravitatoria entre dos cuerpos está descritas por la [**ley de Gravitación Universal**](http://es.wikipedia.org/wiki/ley_de_gravitacion_universal), enunciada por [Sir Isaac Newton](http://es.wikipedia.org/wiki/Isaac_Newton). Dicha ley dice así:


<center>*Dos cuerpos cualesquiera, por el hecho de tener masa, se atraen mutuamente con una fuerza proporcional al producto de sus masas e inversamente proporcional al cuadrado de la distancia que los separa.*</center>


Su expresión matemática es: 

$$F_{1,2} = G\,\frac{m_1 m_2}{r^2}$$

![Esquema](550px-NewtonsLawOfUniversalGravitation.png)
<center>Fuente: *Dennis Nilsson, Crative Commons BY 3.0*</center>

En esta fórmula sabemos qué significa $m_1$ (la masa del cuerpo 1 en kilogramos), $m_2$ (la masa del cuerpo 2, también en kilogramos), y $r$ (la distancia entre los dos cuerpos en metros). Pero... **¿Qué significa $G$?** **¿Qué es?**

$G$ es la llamada [constante de gravitación universal](https://es.wikipedia.org/wiki/constante_de_gravitacion_universal), y su valor en el SI es de $G=6,67\cdot 10^{-11} N \cdot \frac{m^2}{kg^2}$. ¿Por qué tiene este valor tan raro? **Es un misterio**: en el Universo hay ciertas constantes que valen lo que valen sin que podamos encontrar ninguna razón y cuyo valor sólo puede ser obtenido mediante experimentos. Reciben el nombre de [constantes físicas universales](https://es.wikipedia.org/wiki/Constante_f%C3%ADsica).

Eso es **poquísimo**.

## 2. La aceleración de la gravedad

En temas anteriores, habíamos visto que la aceleración de la gravedad es $g=9,81\,m/s$. Esto es algo que nos hemos *aprendido*, que *había que saber*. Vamos a ver ahora de dónde sale esto: Vamos a **deducir el valor de la aceleración de la gravedad**.

Imaginemos la siguiente situación: Fulanito, que *tiene una masa de* $m_{Fulanito}=50\,kg$, está de pie en el patio del recreo. La única fuerza actuando sobre él es la gravedad:

$$P_{Fulanito} = m_{Fulanito}\cdot g$$

Si quisiésemos calcular el peso de Fulanito, multiplicaríamos su masa en kilos por $g=9,81\,m/s$ y listo. Pero vamos a suponer que **no** sabemos cuánto vale $g$. Nos hemos olvidado.

Por suerte, recordamos la ley de la gravitación:

$$P_{Fulanito}=G\, \frac{M_{Tierra}\cdot m_{Fulanito}}{R_{Tierra}^2}= g\cdot m_{Fulanito}$$

Si nos fijamos bien en la ecuación, veremos que:

$$g=G\,\frac{M_{Tierra}}{R_{Tierra}^2}$$.

**¡Y lo mejor de todo es que la masa de Fulanito no interviene para nada!** Todo depende, únicamente, del planeta en el que estemos.

Mirando en [Wikipedia](https://es.wikipedia.org/wiki/Tierra) la masa de la Tierra ($M_{Tierra}$) y su radio ($R_{Tierra}$), podemos ver que:

$$g=6,67\, \left[ N \frac{m^2}{kg} \right] \cdot 10^{-11} \frac{5,9736\cdot 10^{24}\, [kg]}{6371000^2 \, [m^2]}=\textbf{9,81}6276 \, \left[ \frac{m}{s^2} \right]$$

## 3. Movimiento de los planetas: Sistema Sol y Tierra

Nuestro planeta es una gran masa de tierra de forma [aproximadamente esférica](https://es.wikipedia.org/wiki/Geoide), situada en el Espacio y que gira alrededor de una [estrella](https://es.wikipedia.org/wiki/Estrella): El Sol. Dicha estrella tiene una masa muy superior a la de la Tierra y, debido a las leyes de la gravitación y a las [leyes de Kepler](https://es.wikipedia.org/wiki/Leyes_de_Kepler), podemos aproximar el movimiento de la Tierra alrededor del Sol como si fuese un **Moviento Circular Uniforme**, con:

* Período
* Frecuencia
* Fuerzas centrípeta y centrífugas
* Velocidad tangencial

Bien, lo único que necesitamos saber para calcular lo anterior es lo siguiente:

* La Tierra tiene una masa de $5,97\cdot 10^{24}\, kg$
* El Sol tiene una masa de $1,9\cdot 10^{30}\, kg$
* La distancia media Tierra-Sol es de $150000$ millones de $km$
* La Tierra gira alrededor del Sol con **MCU**


```python
import numpy as np
import programa1
from IPython.display import HTML

####################################################################
### Para experimentar con la gravedad y ver cómo cambian         ###
### las órbitas, cambia los valores que aparecen a continuación: ###
####################################################################

# Masa del Sol en kg:
msol = 1.9*10**30
# Masa del planeta en kg:
mplaneta = 5.9*10**24
# Constante de gravitación universal en N m^2/kg:
G = 6.67*10**(-11)
# Distancia entre los dos planetas en m:
distancia = 149597870700


#######################################################################
## Empieza el cálculo: Puedes ver como se hace, pero si cambias algo ##
## es posible que el programa deje de funcionar!                     ##
#######################################################################
# Fuerza de atracción planeta-Sol:
Fgravedad = G*msol*mplaneta/(distancia**2)
# Aceleración centrípeta:
a = Fgravedad/distancia
# velocidad del planeta:
v = np.sqrt(Fgravedad*distancia/mplaneta)
# Período:
T     = 2*np.pi*distancia/v
Tdias = T/3600/24

## Muestro los resultados por pantalla:
programa1.pinta_resultados(Fgravedad,a,v,T,Tdias)

## Muestro la animación por pantalla:
HTML(programa1.mostrar_animacion(a,v,distancia).to_html5_video())
```
